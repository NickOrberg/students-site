# Среды разработки для Python

Мы использовали среду IDLE, она поставляется вместе с питоном и достатчна для того, чтобы начать с ней программировать.

Мы дальше будем использовать среды разработки от компании JetBrains. Это умные среды, которые хорошо понимают код, который вы пишете, подсказыавют возможные ошибки и исправления. Интегрированны с большинством популярных инструментов разработки.

Для python у них есть бесплатная среда PyCharm. Для Java есть IntelliJ IDEA, версия Community Edition бесплатна. Платные версии всех продуктов тоже нам доступны, пока мы учимся и работаем в СПбГУ. Вы можете получить бесплатныю лицензию на любой продукт с помощью университетской почты st....@student.spbu.ru

Что все-таки установить.

1. Самое простое - установите PyCharm
2. IntelliJ IDEA на самом это платформа, к которой можно подкючить любой язык. Java есть там по-умолчанию, но можно туда же добавить и Python, просто доустановив плагин. Это получится точно такая же среда, как в PyCharm.
   * IDEA предназначена для разных языков, поэтому в ней больше меню, кнопок и т.п.
   * Зато IDEA одна позволяет работать и с Java, и с Python
   * в PyCharm новые возможности для работы с Python выходят чуть быстрее.
   
Последнее. Но лучше, наверное, установите их программу Toolbox. Она позволяет
  * управлять IDEA, PyCharm, т.е. открывать их, обновлять.
  * быстро открывать ваши проекты
  * С его помощью легко установить IDEA и PyCharm даже на этих компьютерах
